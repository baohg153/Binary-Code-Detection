## Hướng dẫn sử dụng file `ppl.pkl`
- key: Là chuỗi đã được hash bằng SHA-1 theo công thức:
  ```python
  hashlib.sha1(text.encode("utf-8")).hexdigest()
- value: Giá trị PPL

````python
import pickle
import os
import hashlib
import pandas as pd
PPL_CACHE_PATH = "ppl.pkl"
````

````python
def get_hash(text):
    return hashlib.sha1(text.encode("utf-8")).hexdigest()
````

````python
if os.path.exists(PPL_CACHE_PATH):
    print("loading PPL map from cache...")
    with open(PPL_CACHE_PATH, "rb") as f:
        ppl_map = pickle.load(f)
````

````python
TRAIN_PARQUET = 'train.parquet'
df_train = pd.read_parquet(TRAIN_PARQUET)

def get_hash(text):
    return hashlib.sha1(text.encode("utf-8")).hexdigest()

df_train["perplexity"] = df_train["code"].apply(lambda x: ppl_map.get(get_hash(x)))
````

# EDA

```python
import os
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
TRAIN_PARQUET = '../../data/train.parquet'
VAL_PARQUET = '../../data/validation.parquet'
TEST_SAMPLE_PARQUET = '../../data/test_sample.parquet'
TEST_PARQUET = '../../data/test.parquet'
```

```python
t0 = time.time()
df_train = pd.read_parquet(TRAIN_PARQUET)
df_val   = pd.read_parquet(VAL_PARQUET)
df_ts    = pd.read_parquet(TEST_SAMPLE_PARQUET)
df_test  = pd.read_parquet(TEST_PARQUET)
print(f"loaded all parquets in {time.time()-t0:.1f}s")
```

```python
print(f"train       : {len(df_train):>8,} rows | columns: {list(df_train.columns)[:6]}…")
print(f"val         : {len(df_val):>8,} rows")
print(f"test_sample : {len(df_ts):>8,} rows")
print(f"test        : {len(df_test):>8,} rows")

print("\ntrain language dist       :", df_train['language'].value_counts().astype(int).to_dict())
print("val language dist         :", df_val['language'].value_counts().astype(int).to_dict())
print("test_sample language dist :", df_ts['language'].value_counts().astype(int).to_dict())

print("\ntrain label dist          :", df_train['label'].value_counts().astype(int).to_dict())
print("val label dist            :", df_val['label'].value_counts().astype(int).to_dict())
print("test_sample label dist    :", df_ts['label'].value_counts().astype(int).to_dict())
```

# SAMPLES

```python
human_samples = df_train[df_train['label'] == 0]['code'].head(3)
for i, code in enumerate(human_samples, 1):
    print(f"\n[Sample {i}]")
    print(code)
    print("-" * 30)
```

```python
llm_samples = df_train[df_train['label'] == 1]['code'].head(3)
for i, code in enumerate(llm_samples, 1):
    print(f"\n[Sample {i}]")
    print(code)
    print("-" * 30)
```

```python
def display_samples_by_language(df, label_value, label_name):
    print(f"\n{'='*20} {label_name.upper()} SAMPLES PER LANGUAGE {'='*20}")
    
    subset = df[df['label'] == label_value].groupby('language').first().reset_index()
    
    for _, row in subset.iterrows():
        print(f"\n[Language: {row['language']} | Generator: {row['generator']}]")
        print("-" * 40)
        print(row['code'][:300] + ("..." if len(row['code']) > 300 else ""))
        print("-" * 40)
```

```python
display_samples_by_language(df_train, 0, "Human")
```

```python
display_samples_by_language(df_train, 1, "LLM")
```
